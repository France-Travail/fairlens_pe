# COMPAS Recidivism

The COMPAS dataset consists of the results of a commercial algorithm called COMPAS (Correctional Offender Management 
Profiling for Alternative Sanctions), which was used by judges and parole officers to assess a criminal's likelihood 
of reoffending. In this notebook, we explore the COMPAS dataset for bias toward sensitive features. We go on to show 
similar bias in a logistic regressor trained to forecast a criminal's risk of reoffending using the dataset. [1]

```python
import pandas as pd
import fairlens as fl
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

df = pd.read_csv("https://raw.githubusercontent.com/propublica/compas-analysis/master/compas-scores-two-years.csv")
df
```

## Preprocessing

We drop rows which are not usable according to the analysis done by ProPublica. For more detail see [1].

```python
df = df[(df["days_b_screening_arrest"] <= 30)
        & (df["days_b_screening_arrest"] >= -30)
        & (df["is_recid"] != -1)
        & (df["c_charge_degree"] != 'O')
        & (df["score_text"] != 'N/A')].reset_index(drop=True)

df
```

## Analysis

We start by identifying the legally protected attributes we need to be aware of. The method `fairlens.sensitive.detect_names_df`
can be used to automatically find these attributes in a dataset using fuzzy matching. If the deep search argument is
set to True the method attempts to match all the unique values in a column to potential aliases for protected
attributes, in order to determine if the column is sensitive or not.

```python
sensitive_attributes = fl.sensitive.detect_names_df(df, deep_search=True)
print(sensitive_attributes)
print(sensitive_attributes.keys())
```

We can see that the attributes that we should be concerned about correspond to gender, age and ethnicity.
While FairLens automatically discretizes continuous attributes such as age, the results are often more interpretable
using custom categories; i.e. "Greater than 45", "25 - 45", "Less than 25" in the case of age. Fortunately
the COMPAS dataset comes with a categorical column for age which we can use instead.

```python
df[["sex", "race", "age", "dob", "age_cat", "decile_score"]].head()
```

Now we inspect potential biases in decile scores by visualizing the distributions of different sensitive
sub-groups in the data. Methods in `fairlens.plot` can be used to generate plots of distributions of variables
in different sub-groups in the data.

```python
target_attribute = "decile_score"
sensitive_attributes = ["sex", "race", "age", "dob", "age_cat"]

fl.plot.use_style()
fl.plot.mult_distr_plot(df, target_attribute, sensitive_attributes)
plt.show()
```

The largest horizontal disparity in scores seems to be in race, specifically between African-Americans and Caucasians,
who make up most of the sample. To visualize or measure the distance between just two sub-groups, we provide an optional
interface to simplify writing pandas predicates. For instance `df[df["sex"] == "Male" & df["race"] == "Hispanic"]` 
can be written as `{"sex": ["Male"], "race": ["Hispanic"]}`.

```python
group1 = {"race": ["African-American"]}
group2 = {"race": ["Caucasian"]}

fl.plot.distr_plot(df, "decile_score", [group1, group2])
plt.legend(["African-American", "Caucasian"])
plt.show()
```

The above disparity can be quantified by measuring statistical distances between the two distributions.

```python
from fairlens.metrics import stat_distance

distances = {}
for metric in ["emd", "norm", "hellinger"]:
    distances[metric] = stat_distance(df, "decile_score", group1, group2, mode=metric)


pd.DataFrame.from_dict(distances, orient="index", columns=["distance"])
```

Measuring the statistical distance between the distribution of a variable in a subgroup and in the entire dataset
can indicate how biased the variable is with respect to the subgroup.

```python
fscorer = fl.FairnessScorer(df, "decile_score", ["sex", "race", "age_cat"])
fscorer.distribution_score(max_comb=1, p_value=True)
```

## Training a Model

Now we train a logistic regression model on the COMPAS dataset, to predict a criminal's likelihood of reoffending.

```python
df = df[["sex", "race", "age_cat", "c_charge_degree", "priors_count", "two_year_recid", "score_text"]]
sp = int(len(df) * 0.8)

df_train = df[:sp].reset_index(drop=True)

X = df_train.copy()

X["sex"] = pd.factorize(df_train["sex"])[0]
X["race"] = pd.factorize(df_train["race"])[0]
X["age_cat"].replace(["Greater than 45", "25 - 45", "Less than 25"], [2, 1, 0], inplace=True)
X["c_charge_degree"] = pd.factorize(df_train["c_charge_degree"])[0]
X.drop(columns=["score_text"], inplace=True)
X = X.to_numpy()

y = pd.factorize(df_train["score_text"] != "Low")[0]

print(X.shape)
print(y.shape)
```

Below we see the distributions of the predictions on the training data is similar to the distributions of decile
score in the dataset as expected.

```python
clf = LogisticRegression(random_state=0).fit(X, y)
df_train["pred"] = clf.predict_proba(X)[:, 1]

fscorer = fl.FairnessScorer(df_train, "pred", ["race", "sex", "age_cat"])
fscorer.plot_distributions()
plt.show()
```

We now show similar results on a held out test set.

```python
df_test = df[sp:].reset_index(drop=True)

X = df_test.copy()

X["sex"] = pd.factorize(df_test["sex"])[0]
X["race"] = pd.factorize(df_test["race"])[0]
X["age_cat"].replace(["Greater than 45", "25 - 45", "Less than 25"], [2, 1, 0], inplace=True)
X["c_charge_degree"] = pd.factorize(df_test["c_charge_degree"])[0]
X.drop(columns=["score_text"], inplace=True)
X = X.to_numpy()

print(X.shape)
```

```python
df_test["pred"] = clf.predict_proba(X)[:, 1]

fscorer = fl.FairnessScorer(df_test, "pred", ["race", "sex", "age_cat"])
fscorer.plot_distributions()
plt.show()
```

```python
fscorer.distribution_score(max_comb=1, p_value=True).sort_values("Distance", ascending=False).reset_index(drop=True)
```

Lets try dropping "race" from the model and look at the results.

```python
X = df_train[["c_charge_degree", "sex", "age_cat", "priors_count", "two_year_recid", "score_text"]].copy()
X["sex"] = pd.factorize(df_train["sex"])[0]
X["age_cat"].replace(["Greater than 45", "25 - 45", "Less than 25"], [2, 1, 0], inplace=True)
X["c_charge_degree"] = pd.factorize(df_train["c_charge_degree"])[0]
X.drop(columns=["score_text"], inplace=True)
X = X.to_numpy()

y = pd.factorize(df_train["score_text"] != "Low")[0]

print(X.shape)
print(y.shape)
```

```python
clf = LogisticRegression(random_state=0).fit(X, y)
df_train["pred_dropped"] = clf.predict_proba(X)[:, 1]

fscorer = fl.FairnessScorer(df_train, "pred_dropped", ["race", "sex", "age_cat"])
fscorer.plot_distributions()
plt.show()
```

```python
X = df_test[["c_charge_degree", "sex", "age_cat", "priors_count", "two_year_recid", "score_text"]].copy()

X["sex"] = pd.factorize(df_test["sex"])[0]
X["age_cat"].replace(["Greater than 45", "25 - 45", "Less than 25"], [2, 1, 0], inplace=True)
X["c_charge_degree"] = pd.factorize(df_test["c_charge_degree"])[0]
X.drop(columns=["score_text"], inplace=True)
X = X.to_numpy()

print(X.shape)
```

```python
df_test["pred_dropped"] = clf.predict_proba(X)[:, 1]

fscorer = fl.FairnessScorer(df_test, "pred_dropped", ["race", "sex", "age_cat"])
fscorer.plot_distributions()
plt.show()
```

```python
fscorer.distribution_score(max_comb=1, p_value=True).sort_values("Distance", ascending=False).reset_index(drop=True)
```

We can see that despite dropping the attribute "race", the biases toward people in different racial groups is almost
unaffected. This is because, although we have dropped the attribute, there are hidden correlations between the column
"race" and columns which inform the decision of the model.
